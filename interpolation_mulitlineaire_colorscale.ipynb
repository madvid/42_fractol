```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mako_palette = sns.color_palette("mako", as_cmap=True)
viridis_palette = sns.color_palette("viridis", as_cmap=True)
magma_palette = sns.color_palette("magma", as_cmap=True)
ylorbr_palette = sns.color_palette("YlOrBr", as_cmap=True)
coolwarm_palette = sns.color_palette("coolwarm", as_cmap=True)
red_palette = sns.color_palette("Reds", as_cmap=True)
blue_palette = sns.color_palette("Blues", as_cmap=True)
green_palette = sns.color_palette("Greens", as_cmap=True)
```

```python
mako_palette_RGB = [(mako_palette(int(i))[0]*255, mako_palette(int(i))[1]*255, mako_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
viridis_palette_RGB = [(viridis_palette(int(i))[0]*255, viridis_palette(int(i))[1]*255, viridis_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
magma_palette_RGB = [(magma_palette(int(i))[0]*255, magma_palette(int(i))[1]*255, magma_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
ylorbr_palette_RGB = [(ylorbr_palette(int(i))[0]*255, ylorbr_palette(int(i))[1]*255, ylorbr_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
coolwarm_palette_RGB = [(coolwarm_palette(int(i))[0]*255, coolwarm_palette(int(i))[1]*255, coolwarm_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
red_palette_RGB = [(red_palette(int(i))[0]*255, red_palette(int(i))[1]*255, red_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
blue_palette_RGB = [(blue_palette(int(i))[0]*255, blue_palette(int(i))[1]*255, blue_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
green_palette_RGB = [(green_palette(int(i))[0]*255, green_palette(int(i))[1]*255, green_palette(int(i))[2]*255) for i in np.linspace(0,255, 100)]
```

```python
def convert_dataframe(palette):
    l_red = []
    l_blue = []
    l_green = []
    for elem in palette:
        l_red.append(elem[0])
        l_blue.append(elem[1])
        l_green.append(elem[2])
    df = pd.DataFrame({"red":l_red, "blue":l_blue, "green":l_green})
    return (df)
```

```python
df_mako = convert_dataframe(mako_palette_RGB)
df_viridis = convert_dataframe(viridis_palette_RGB)
df_magma = convert_dataframe(magma_palette_RGB)
df_ylorbr = convert_dataframe(ylorbr_palette_RGB)
df_coolwarm = convert_dataframe(coolwarm_palette_RGB)
df_red = convert_dataframe(red_palette_RGB)
df_blue = convert_dataframe(blue_palette_RGB)
df_green = convert_dataframe(green_palette_RGB)
```

```python
fig, axes = plt.subplots(3,3, figsize=(20,20))

axes[0][0].plot(df_mako["red"], label="red", color="red")
axes[0][0].plot(df_mako["blue"], label="blue", color="blue")
axes[0][0].plot(df_mako["green"], label="green", color="green")
axes[0][0].set_title("color palette mako")

axes[0][1].plot(df_viridis["red"], label="red", color="red")
axes[0][1].plot(df_viridis["blue"], label="blue", color="blue")
axes[0][1].plot(df_viridis["green"], label="green", color="green")
axes[0][1].set_title("color palette viridis")

axes[0][2].plot(df_magma["red"], label="red", color="red")
axes[0][2].plot(df_magma["blue"], label="blue", color="blue")
axes[0][2].plot(df_magma["green"], label="green", color="green")
axes[0][2].set_title("color palette magma")

axes[1][0].plot(df_ylorbr["red"], label="red", color="red")
axes[1][0].plot(df_ylorbr["blue"], label="blue", color="blue")
axes[1][0].plot(df_ylorbr["green"], label="green", color="green")
axes[1][0].set_title("color palette ylorbr")

axes[1][1].plot(df_coolwarm["red"], label="red", color="red")
axes[1][1].plot(df_coolwarm["blue"], label="blue", color="blue")
axes[1][1].plot(df_coolwarm["green"], label="green", color="green")
axes[1][1].set_title("color palette coolwarm")

axes[1][2].plot(df_red["red"], label="red", color="red")
axes[1][2].plot(df_red["blue"], label="blue", color="blue")
axes[1][2].plot(df_red["green"], label="green", color="green")
axes[1][2].set_title("color palette red")

axes[2][0].plot(df_blue["red"], label="red", color="red")
axes[2][0].plot(df_blue["blue"], label="blue", color="blue")
axes[2][0].plot(df_blue["green"], label="green", color="green")
axes[2][0].set_title("color palette blue")

axes[2][1].plot(df_green["red"], label="red", color="red")
axes[2][1].plot(df_green["blue"], label="blue", color="blue")
axes[2][1].plot(df_green["green"], label="green", color="green")
axes[2][1].set_title("color palette green")
```
